### Merging Cases, deaths and population datasets using countyFIPS

```python
merged_data = pd.merge(cases, deaths, on= ['countyFIPS','County Name', 'State','StateFIPS'], suffixes=('_cases', '_deaths'))
merged_data.head()
merged_data.shape
```

```python
final_merged_data =pd.merge(merged_data,population, on= ['countyFIPS'])
final_merged_data
```

```python
final_merged_data.head()
```

```python

columns_to_drop =["County Name_y", "State_y"]
final_merged_data.drop(columns=columns_to_drop, inplace=True)
```

```python
column_rename = {
    'County Name_x': 'County Name',
    'State_x': 'State'
}

# Rename the columns with suffixes
final_merged_data.rename(columns=column_rename, inplace=True)
```
