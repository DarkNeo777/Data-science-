### Import COVID 19 data 

```python
import pandas as pd
import matplotlib.pyplot as plt
pd.reset_option('display.max_rows')
#pd.set_option('display.max_rows', None)
```

```python
deaths = pd.read_csv("data/covid_deaths_usafacts.csv")
cases = pd.read_csv("data/covid_confirmed_usafacts.csv")
population = pd.read_csv("data/covid_county_population_usafacts.csv")
```

#### Merging data to one dataset

```python
merged_data = pd.merge(cases, deaths, on= ['countyFIPS','County Name', 'State','StateFIPS'], suffixes=('_cases', '_deaths'))
final_data = pd.merge(merged_data, population, on=['countyFIPS', 'State'])
```

#### Cleaning up the data

```python

final_data.columns
```

```python
final_data.drop('County Name_y',axis=1, inplace=True)
```

```python

final_data = final_data.rename(columns={"County Name_x": "County Name"})
```

```python
population = final_data.pop('population')
final_data.insert(4, 'population', population)
```

```python
final_data
```

# Enrichment data

```python
QCEW = pd.read_csv("data/Quarterly Census of Employment and Wages.csv", low_memory=False)
```

```python
QCEW.head()
```

## Cleaning the enrichment data:

```python
new_QCEW = QCEW.dropna()
print(QCEW.columns)
```

```python
print(new_QCEW.columns)
```

#### Removing unwanted columns

```python
columns_to_remove = ['St', 'Cnty', 'Own', 'NAICS', 'Year', 'Qtr', 'Area Type',
        'Status Code', 'January Employment', 'February Employment',
       'March Employment', 'Average Weekly Wage',
       'Employment Location Quotient Relative to U.S.',
       'Total Wage Location Quotient Relative to U.S.']

clean_data = new_QCEW.drop(columns_to_remove, axis=1)
clean_data
```

```python
clean_data.dtypes
```

```python
print("There are", clean_data.shape[0],"of Observations and ", clean_data.shape[1], " features in cases dataset.")
print("\n")
```

```python
print(clean_data.isnull().sum())
print("\n")
```

```python
clean_data = clean_data.rename(columns={'St Name': 'State'})
```

```python
clean_data = clean_data.rename(columns={'Area\nCode': 'countyFIPS'})
clean_data
```

```python
clean_data['countyFIPS'] = clean_data['countyFIPS'].astype(int)
```

## Merging the enrichment data with covid dataset

```python
final_complete_data = final_data.merge(clean_data, on=['countyFIPS'], how='left')
```
