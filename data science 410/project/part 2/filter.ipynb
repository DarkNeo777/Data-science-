```python
import numpy as np
import pandas as pd

file_path = 'data/merged_data.csv'  # Replace with your file path
df = pd.read_csv(file_path)
df.head()
```

```python
df = df.drop(['countyFIPS', 'population'], axis=1)
case_columns = [col for col in df.columns if '_cases' in col]
death_columns = [col for col in df.columns if '_deaths' in col]
df_cases = df.melt(id_vars=['County Name', 'State', 'StateFIPS'], value_vars=case_columns, 
                   var_name='date', value_name='cases')
df_cases['date'] = df_cases['date'].str.replace('_cases', '')

df_deaths = df.melt(id_vars=['County Name', 'State', 'StateFIPS'], value_vars=death_columns, 
                    var_name='date', value_name='deaths')
df_deaths['date'] = df_deaths['date'].str.replace('_deaths', '')
```

```python
df_combined = pd.merge(df_cases, df_deaths, on=['County Name', 'State', 'StateFIPS', 'date'])
df_grouped = df_combined.groupby(['State', 'date']).agg({'cases': 'sum', 'deaths': 'sum'}).reset_index()
```

```python
df_grouped['date'] = pd.to_datetime(df_grouped['date'])
start_date = '2020-06-28'
end_date = '2020-12-27'

mask = (df_grouped['date'] >= start_date) & (df_grouped['date'] <= end_date)
filtered_df = df_grouped.loc[mask]
```

```python
filtered_df.set_index('date', inplace=True)
filtered_df.head()
```

```python
# Replace 'StateName' with the name of the state you want to filter by
NC_df = filtered_df.loc[filtered_df['State'] == 'NC']
NC_df.head()
```
