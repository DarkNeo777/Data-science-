*** 4. Create age categories: children (under 14 years), adolescents (14-20), adult (21-64), and senior(65+), and calculate survival proportions by age category, pclass and sex. (30 pts)***

```python
import pandas as pd

# Define the age category bins and labels
age_bins = [0, 14, 21, 65, float('inf')]
age_labels = ['children', 'adolescents', 'adult', 'senior']

# Create a new column 'AgeCategory' based on the age bins
titanic_d['Age_Group'] = pd.cut(titanic_d['age'], bins=age_bins, labels=age_labels)

# Group data by Age Group, Pclass, and Sex and calculate the mean of the sruvived
survival_proportions = titanic_d.groupby(['Age_Group', 'pclass', 'sex'])['survived'].mean().unstack()

survival_proportions
```
